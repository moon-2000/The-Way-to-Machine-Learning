### Import required libraries

```python
import pandas as pd
import numpy as np

from sklearn import preprocessing
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
sns.set()

import warnings
warnings.filterwarnings('ignore')
```

### Load Data

```python
df=pd.read_csv("data_logReg.csv")
df.head()
```

### Explore the data

```python
# get the type of each column
df.dtypes
```

```python
# get the dimensions of the data
df.shape
```

```python
# check missing values
df.isnull().sum()
```

```python
# check for duplicate values
duplicate_rows_df = df[df.duplicated()]

print(f"Number of duplicate rows: {duplicate_rows_df.shape}")
```

### Preprocess the data

```python
# drop the loan ID column from the dataset
df=df.drop(["Loan_ID"],axis=1)
df.shape
```

```python
# drop the rows that have missing values
df = df.dropna()
df.shape
```

```python
# encode the categorical variables into numerical ones
le = preprocessing.LabelEncoder()
df[df.columns[0]]=le.fit_transform(df.iloc[:,0])
df[df.columns[1]]=le.fit_transform(df.iloc[:,1])
df[df.columns[2]]=le.fit_transform(df.iloc[:,2])
df[df.columns[3]]=le.fit_transform(df.iloc[:,3])
df[df.columns[4]]=le.fit_transform(df.iloc[:,4])
df[df.columns[10]]=le.fit_transform(df.iloc[:,10])
df[df.columns[11]]=le.fit_transform(df.iloc[:,11])
```

```python
# check the distribution of the target classes 1 and 0
df.Loan_Status.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("loan pediction distribution")
plt.ylabel("customers")
plt.xlabel("target")
```

```python
from imblearn.over_sampling import SMOTE
os = SMOTE(random_state=0)
```

```python
df.head()
```

```python
X = df.loc[:, df.columns != 'Loan_Status']  # store all features in X 
y = df.loc[:, df.columns == 'Loan_Status']  # store the target column in y
y
```

### Oversampling with SMOTE

```python
from imblearn.over_sampling import SMOTE
os = SMOTE(random_state=0)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
columns = X_train.columns
os_data_X, os_data_y = os.fit_resample(X_train, y_train)

os_data_X = pd.DataFrame(data=os_data_X,columns=columns )
os_data_y= pd.DataFrame(data=os_data_y,columns=['Loan_Status'])
os_data_y
```

```python
# we can Check the numbers of our data
print("length of oversampled data is ",len(os_data_X))
print("Number of no subscription in oversampled data",len(os_data_y[os_data_y['Loan_Status']==0]))
print("Number of subscription",len(os_data_y[os_data_y['Loan_Status']==1]))
print("Proportion of no subscription data in oversampled data is ",len(os_data_y[os_data_y['Loan_Status']==0])/len(os_data_X))
print("Proportion of subscription data in oversampled data is ",len(os_data_y[os_data_y['Loan_Status']==1])/len(os_data_X))
```

### Scale the Features

```python
# Example of using MinMaxScaler for normalization
from sklearn.preprocessing import MinMaxScaler

# Instantiate the scaler
scaler = MinMaxScaler()

# Scale the input features
os_data_X_scaled = scaler.fit_transform(os_data_X)
```

### Build the model

```python
# Fit the logistic regression model with scaled data
logReg = LogisticRegression()

logReg.fit(os_data_X_scaled, os_data_y.values.ravel())
```

```python
y_pred = logReg.predict(X_test)

print('Accuracy of logistic regression classifier on test set: {:.3f}'.format(logReg.score(X_test, y_test)))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
