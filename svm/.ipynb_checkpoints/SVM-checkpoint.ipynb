### Define the dependent and independent variables

```python
X = df.drop(['target_class'], axis=1)

y = df['target_class']
```

### Prepare the data for the model

```python
# split X and y into training and testing sets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

#### Feature Scaling

```python
cols = X_train.columns

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)

X_test = scaler.transform(X_test)
```

```python
X_train = pd.DataFrame(X_train, columns=[cols])

X_test = pd.DataFrame(X_test, columns=[cols])
```

```python
X_train.describe()
```

### Run an SVM model with default hyperparameters

```python
# import SVC classifier
from sklearn.svm import SVC

# import metrics to compute accuracy
from sklearn.metrics import accuracy_score
```

```python
# create an instance of SVC classifier with default hyperparameters
svc=SVC() 

# fit the classifier to training set
svc.fit(X_train,y_train)
```

```python
# make predictions on test set
y_pred=svc.predict(X_test)

# compute and print accuracy score
print('The accuracy of this SVC model with default hyperparameters: {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
```

### Using sigmoid kernel

```python
# Create an instance of SVC classifier with a "sigmoid" kernel while C = 1.0.
sigmoid_svc=SVC(kernel='sigmoid', C=1.0) 

# fit classifier to training set
sigmoid_svc.fit(X_train,y_train)
```

```python
# make predictions on test set
y_pred=sigmoid_svc.predict(X_test)

# compute and print accuracy score
print('Model accuracy score with sigmoid kernel and C=1.0 : {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
```

### Using linear kernal

```python
# Create an instance of SVC classifier with a "linear" kernel while C = 1.0.
linear_svc=SVC(kernel='linear', C=1.0) 

# fit classifier to training set
linear_svc.fit(X_train,y_train)
```

```python
# make predictions on test set
y_pred_test=linear_svc.predict(X_test)

# compute and print accuracy score
print('Model accuracy score with linear kernel and C=1.0 : {0:0.4f}'. format(accuracy_score(y_test, y_pred_test)))
```

#### Change C value

```python
# Create an instance of SVC classifier with a "linear" kernel while C = 100.0.
linear_svc100=SVC(kernel='linear', C=100.0) 

# fit classifier to training set
linear_svc100.fit(X_train, y_train)
```

```python
# make predictions on test set
y_pred=linear_svc100.predict(X_test)

# compute and print accuracy score
print('Model accuracy score with linear kernel and C=100.0 : {0:0.4f}'. format(accuracy_score(y_test, y_pred)))
```
